```python
import cv2
from matplotlib import pyplot as plt
```

```python
imageFile='./lenna.bmp'
imgGray=cv2.imread(imageFile,cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(6,6))
plt.subplots_adjust(left=0,right=1,bottom=0,top=1)#영상 출력을 좌우 [0,1] 위아래를 [0,1]
plt.imshow(imgGray,cmap='gray')
```

```python
plt.axis('off')#축을 없앴다
plt.imshow(imgGray,cmap='gray')
```
