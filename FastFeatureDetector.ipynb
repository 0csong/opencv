```python
import cv2
import numpy as np
from IPython.display import Image
```

```python
src = cv2.imread('./data/chessBoard.jpg')
gray= cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
Image('./data/chessBoard.jpg')
```

```python
fastF =cv2.FastFeatureDetector.create(threshold=30)#T가 30인 fastF생성
#T값이 100일시에 특징점은 줄어듬
kp = fastF.detect(gray) #gray로 특징점 추출
dst = cv2.drawKeypoints(gray, kp, None, color=(0,0,255))#특징점 표시
print('len(kp)=', len(kp))#특징점 98개
cv2.imshow('dst',  dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
fastF.setNonmaxSuppression(False)#지역극값 억제 x
kp2 = fastF.detect(gray)#gray로 특징점 추출
dst2 = cv2.drawKeypoints(src, kp2, None, color=(0,0,255))
print('len(kp2)=', len(kp2))#특징점 867개
cv2.imshow('dst2',  dst2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
dst3 = src.copy()
points = cv2.KeyPoint_convert(kp)
for cx, cy in points:
    cv2.circle(dst3, (int(cx), int(cy)), 3, color=(255, 0, 0), thickness=1)
cv2.imshow('dst3',  dst3)
cv2.waitKey()
cv2.destroyAllWindows()
```
